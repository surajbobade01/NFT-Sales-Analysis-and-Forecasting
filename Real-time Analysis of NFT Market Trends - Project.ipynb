```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the Data

```python
# Convert date to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set the date column as the Index
df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(14,7)) 
plt.plot(df.index, df['Sales_USD_cumsum'], color='g')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales (USD)')
plt.title('Cumulative NFT Sales Over Time')
plt.grid(True)
plt.show() # Plot cumulative sales over time
```

```python
plt.figure(figsize=(14, 7))
plt.hist(df['Sales_USD'].dropna(), bins=50, color='g')
plt.xlabel('Sales Price (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of NFT Sales Prices')
plt.grid(True)
plt.show()  # Plot the distribution of sales prices
```

```python
top_10_sales = df.sort_values(by='Sales_USD', ascending=False).head(10)
print(top_10_sales[['Sales_USD', 'Number_of_Sales', 'Active_Market_Wallets', 'Primary_Sales', 'Secondary_Sales_cumsum']])  # Top 10 sales by price
```
